```python
# 인스턴스 속성 사용

class Person():
    def __init__(self):
        self.hello = '안녕하세요' # 인스턴스 변수

    def hi(self):
        print(self.hello)

jin = Person() # 이 때 __init__ 메서드는 실행됐다.
jin.hi()
```

```python
class Person():
    def __init__(self, name, age, hobby):
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.hobby = hobby

    def introduce(self):
        print(f"{self.hello}, 저는 {self.age}살 {self.name}입니다") # 클래스 내부에서 접근

jin = Person('진', 22, '운동')
jin.introduce()

paul = Person('폴', 22, '요리')
paul.introduce()

print(f"이름 : {jin.name}") # 클래스 외부에서 접근
print(f"나이 : {jin.age}")
print(f"취미 : {jin.hobby}")
```

```python
class Person:
    pass


jin = Person() # 인스턴스 생성
jin.name = '진' # 인스턴스를 생성한 뒤 속성 추가

paul = Person()
paul.name = '폴'

print(jin.name)
print(paul.name)
```

```python
# 클래스 외부에서는 접근 불가,
# 클래스 내부에서만 사용 가능한 private attribute

class Person():
    def __init__(self, name, age, hobby, weight):
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.hobby = hobby
        self.__weight = weight # 비공개 속성

    def diet(self, loss_weight):
        self.__weight -= loss_weight
        print(self.__weight)

jin = Person('진', 22, '운동', 88)
jin.diet(10)
```
