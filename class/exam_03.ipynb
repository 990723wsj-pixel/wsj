```python
# 클래스 속성

class Person:
    folder = [] # 클래스 속성(공공제) -> 
    # 클래스에 속한다. 모든 인스턴스에서 접근이 가능하다.

    def add_files(self, file):
        self.folder.append(file) # 클래스 이름으로 클래스 속성에 접근할 수 있다.

jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')

print(jin.folder)
print(paul.folder)
```

```python
# 인스턴스 속성

class Person:
    def __init__(self):
        self.folder = [] # 인스턴스 속성

    def add_files(self, file):
        self.folder.append(file) # 클래스 이름으로 클래스 속성에 접근할 수 있다.

jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')

print(jin.folder)
print(paul.folder)
```

```python
# 비공개 클래스 속성

class Person:
    __weight = 78

    def print_weight(self):
        print(Person.__weight)

jin = Person()
jin.print_weight()

# print(jin.__weight)

# 인스턴스 속성
# -> 각각의 객체가 가지는 속성

# 클래스 속성
# -> 속성이 클래스에 딱 하나 존재하기 때문에 공유가 된다.
# -> 클래스에 하나만 존재하며 모든 인스턴스가 공유하는 속성
```
