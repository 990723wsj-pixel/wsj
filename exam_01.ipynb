```python
#문제 - 입력받은 정수가 소수인지 아닌지 알려주는 함수를 구현해주세요.(개념 : 함수, 반복문, 리턴)
def is_prime_number(num):
    if num <=1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
        return True
print("3은 소수입니다 : {}".format(is_prime_number(3)))
print("4는 소수입니다 : {}".format(is_prime_number(4)))
print("5는 소수입니다 : {}".format(is_prime_number(5)))
print("6는 소수입니다 : {}".format(is_prime_number(6)))
print("7는 소수입니다 : {}".format(is_prime_number(7)))
print("1000은 소수입니다 : {}".format(is_prime_number(1000)))
```

```python
def is_prime_number(num):
    if num <=1:
        return False
    i = 2
    while i < num:
        if num % i == 0:
            return False
        i += 1

    return True


print("3은 소수입니다 : {}".format(is_prime_number(3)))
print("4는 소수입니다 : {}".format(is_prime_number(4)))
print("5는 소수입니다 : {}".format(is_prime_number(5)))
print("6는 소수입니다 : {}".format(is_prime_number(6)))
print("7는 소수입니다 : {}".format(is_prime_number(7)))
print("1000은 소수입니다 : {}".format(is_prime_number(1000)))
```

```python
#문제 - 딕셔너리에 각 달의 마지막 날들을 한번에 담기
#문제 : 딕셔너리에 각 달의 마지막 날들을 한번에 담기
#조건 : 1월 ~ 3월까지
#조건 : 2월의 마지막은 28일로 함
#조건 : 첫 데이터의 key는 "1월", value는 31 
month_end_days= {"1월":31, "2월":28, "3월":31}
print(month_end_days)
```

```python
print("== 정답 v1 ==")
month_end_days = {}

month_end_days["1월"] = 31
month_end_days["2월"] = 28
month_end_days["3월"] = 31

print(month_end_days)

print("== 정답 v2 ==")
month_end_days = {}

for i in range(1, 12 + 1):
  month = str(i) + "월"

  end_day = 31
  if i == 2:
    end_day = 28
  elif i in [4, 6, 9, 11]:
    end_day = 30
  month_end_days[month] = end_day
print(month_end_days)
```

```python
month_end_days = {}

month_end_days_list_ver = [31,28,31,30,31,30,31,31,30,31,30,31]

for i, end_days in enumerate(month_end_days_list_ver):
    month = str(i+1) + "월"
    month_end_days[month] = end_days

print(month_end_days)
```
